> ## ノック13: データに揺れがあるまま集計してみよう

```
df_uriage['purchase_date'] = pd.to_datetime(df_uriage['purchase_date'])
df_uriage['purchase_month'] = df_uriage['purchase_date'].dt.strftime('%Y%m')
```

```
df_uriage.pivot_table(index = 'purchase_month', columns = 'item_name', aggfunc = np.sum, fill_value=0)
```

> ## ノック14: 商品名の揺れを補正しよう

```
df_uriage['item_name'].unique()
```

```
df_uriage['item_name'] = df_uriage['item_name'].str.replace(' ', '')
df_uriage['item_name'] = df_uriage['item_name'].str.upper()
```

```
df_uriage['item_name'].unique()
```

> ## ノック15: 金額の欠損値の補完

```
df_uriage['item_price'].head()
```

商品ごとの金額で欠損を埋める.

```
df_uriage['item_price'] = df_uriage.groupby('item_name')['item_price'].transform(lambda x: x.fillna(x.mean()))
df_uriage.head()
```

```
df_uriage.isnull().any(axis = 0)
```

> ## ノック6: 顧客名の揺れを補正しよう

```
df_kokyaku_daicho['顧客名'].unique()
```

```
df_kokyaku_daicho['顧客名'] = df_kokyaku_daicho['顧客名'].str.replace("  ", "")
df_kokyaku_daicho['顧客名'] = df_kokyaku_daicho['顧客名'].str.replace(" ", "")
df_kokyaku_daicho['顧客名'] = df_kokyaku_daicho['顧客名'].str.replace('\u3000', '')
```

```
df_kokyaku_daicho['顧客名'].unique()
```

> ## ノック17: 日付の揺れを補正しよう

```
df_kokyaku_daicho['登録日'].head()
```

```
date_flg = df_kokyaku_daicho['登録日'].astype('str').str.isdigit()
date_corrected = pd.to_timedelta(df_kokyaku_daicho.loc[date_flg, '登録日'].astype('float') -2  , unit="D") + pd.to_datetime( "1900/01/01") 
date_corrected
```

メモ：
- excel のシリアル列は2日ずれとなっている (p.64 参照)

```
 df_kokyaku_daicho['登録日'].astype('str').str.isdigit().sum()
```

```
df_kokyaku_daicho.loc[date_flg, '登録日'] = date_corrected
df_kokyaku_daicho['登録日'] = pd.to_datetime(df_kokyaku_daicho['登録日'])
```

全て補正されたことを確認.

```
# 登録年月
df_kokyaku_daicho['登録年月'] = df_kokyaku_daicho['登録日'].dt.strftime("%Y%m")
df_kokyaku_daicho.groupby('登録年月').count()['顧客名']
```

> ## ノック18: 顧客名をキーに2つのデータを結合しよう



```
df_kokyaku_daicho.head(3)
```

```
df_all = pd.merge(df_uriage, df_kokyaku_daicho, left_on='customer_name', right_on='顧客名', how='left')
df_all.head()
```

```
# 加工済みデータを保存しておく
dump_data = df_all.copy()
dump_data.to_csv(PATH + 'dump_data.csv', index=False)
```
